```
import matplotlib.pyplot as plt
import math
import matplotlib.font_manager as fm
from tkinter import *
from matplotlib import animation
import numpy as np

get_ipython().run_line_magic('matplotlib', 'auto')

root=Tk()
root.title("GUI")
root.geometry("640x500")

def backtomain():
    global l,f,j,label1,label2,label3,btnRead,btn,listbox
    l.pack_forget()
    f.pack_forget()
    j.pack_forget()
    label1.pack_forget()
    label2.pack_forget()
    label3.pack_forget()
    btnRead.pack_forget()
    btn.pack_forget()
    listbox.pack_forget()
    btn1.pack()
    btn2.pack()
    
    pass

def getTextInput1():
    global l,f,j
    l_=float(l.get("1.0","end"))
    f_=float(f.get("1.0","end"))
    j_=float(j.get("1.0","end"))
    return(ibsa(l_,f_,j_))

def getTextInput2():
    global l,f,j
    l_=float(l.get("1.0","end"))
    f_=float(f.get("1.0","end"))
    j_=float(j.get("1.0","end"))
    return(padong(l_,f_,j_))
    
def inp1():
    btn1.pack_forget()
    btn2.pack_forget()
    global l,f,j,label1,label2,label3,btnRead,btn,listbox
    listbox = Listbox(root,selectmode="extended",height=0)
    listbox.insert(0,"진공:1")
    listbox.insert(1,"물:1.33")
    listbox.insert(2,"유리:1.52")
    listbox.insert(END,"다이아몬드:2.42")
    listbox.insert(END,"공기:1")
    listbox.insert(END,"벤젠:1.50")
    listbox.pack(side='right',anchor='ne')
    label1=Label(root, text="입사되는 쪽의 매질의 굴절률 n1을 입력하세요")
    label1.pack()
    l=Text(root, height=3)
    l.pack()
    label2=Label(root, text="입사되는 반대편의 매질의 굴절률 n2를 입력하세요")
    label2.pack()
    f=Text(root, height=3)
    f.pack()
    label3=Label(root, text="입사각 i(º단위)를 입력하세요")
    label3.pack()
    j=Text(root, height=3)
    j.pack()
    btnRead=Button(root, height=1, width=10, text="그래프 생성", command=getTextInput1)
    btnRead.pack()
    btn=Button(root,text="메인 화면",command=backtomain)
    btn.pack()

    
def inp2():
    global l,f,j,label1,label2,label3,btnRead,btn,listbox
    btn1.pack_forget()
    btn2.pack_forget()
    listbox = Listbox(root,selectmode="extended",height=0)
    listbox.pack
    label1=Label(root, text="파동의 파장을 입력하세요")
    label1.pack()
    l=Text(root, height=3)
    l.pack()
    label2=Label(root, text="파동의 진동수(Hz)를 입력하세요(진동수=1/주기)")
    label2.pack()
    f=Text(root, height=3)
    f.pack()
    label3=Label(root, text="파장의 진폭을 입력하세요")
    label3.pack()
    j=Text(root, height=3)
    j.pack()
    btnRead=Button(root, height=1, width=10, text="그래프 생성", command=getTextInput2)
    btnRead.pack()
    btn=Button(root,text="메인 화면",command=backtomain)
    btn.pack()
    
def ibsa(n1,n2,i):

    title_font_name = 'HYPost'
 
    title_font_path = [f.fname for f in fm.fontManager.ttflist if title_font_name in f.name][0]
 
    title_font_size = 20
 
    title_font_prop = fm.FontProperties(fname = title_font_path, size=title_font_size)
    
    if(round(n1*math.sin(i*math.pi/180 )/n2,4)>1):
        r=-round(i*math.pi/180,4)

    else:
        r=round(math.asin(n1*math.sin(i*math.pi/180 )/n2),4)

    x1 = range(-100, 0)
    x2= range(0,100)
    y1 = [v/math.tan(i*math.pi/180) for v in x1]
    y2 = [v/math.tan(r) for v in x2]

    x=range(-100,100)
    y=y1+y2
    z=[0 for n in x]

    plt.figure(figsize = (6,6))
    plt.axis([-100,100,-100,100])
    plt.axis('off')
    plt.text(-90,5, 'n2')
    plt.text(-90, -10, 'n1')
    if(n1*math.sin(i*math.pi/180 )/n2>1):
        plt.text(80,10, '전반사',fontproperties=title_font_prop)
        pass
    elif(round(math.sin(i*math.pi/180),4)==n2/n1):
        plt.text(80,10, '임계각', fontproperties=title_font_prop)
        pass
    else:
        plt.text(80,10, '굴절',fontproperties=title_font_prop)

    plt.plot(x, y, x, z)
    plt.show()
    

def padong(l,f,j):
    fig = plt.figure()
    ax = plt.axes(xlim=(0, 2), ylim=(-5-j, 5+j))
    lines = ax.plot([], [], lw=2)
    dots =ax.plot([1],[0],'ro', markersize=4)
    patches=lines+dots

    def init():
        for line in lines:
            line.set_data([], [])
            
        for dot in dots:
            dot.set_data(1,0)
        return patches


    def animate(i):
        x = np.linspace(0, 2, 1000)
        y = j*np.sin(12/l * np.pi * (x - 0.01 * i))
        for line in lines:
            line.set_data(x,y)
            
        for dot in dots:
            dot.set_data(1,j*np.sin(12/l * np.pi * (1- 0.01 * i)))
        return patches
    
    anim = animation.FuncAnimation(fig, animate, init_func=init,frames=200, interval=180.0/(l*f), blit=True)
    plt.show()
    
title=Label(root, text="파동의 시각화", height=2,font=("",20))
title.pack()
btn1=Button(root,text="빛의 입사&전반사",command=inp1)
btn1.pack()
btn2=Button(root,text="파동의 형성",command=inp2)
btn2.pack()
root.mainloop()
```
